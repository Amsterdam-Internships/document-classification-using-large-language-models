```python
import pandas as pd

# Create a simple dataset
data = {
    'text': [
        'I love programming.',
        'I hate bugs.',
        'Python is amazing.',
        'Debugging is fun.',
        'I enjoy learning new things.'
    ],
    'label': [1, 0, 1, 1, 1]
}

df = pd.DataFrame(data)
```

```python
from transformers import BertTokenizer

tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')

def preprocess_data(text, tokenizer, max_len):
    encoding = tokenizer.encode_plus(
        text,
        add_special_tokens=True,
        max_length=max_len,
        return_token_type_ids=False,
        padding='max_length',
        truncation=True,
        return_attention_mask=True,
        return_tensors='pt',
    )
    return encoding

# Example of preprocessing
max_len = 20
encoded_data = [preprocess_data(text, tokenizer, max_len) for text in df['text']]
```
